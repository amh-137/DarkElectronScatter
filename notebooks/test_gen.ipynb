```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
particle_list = []
for i in range(100000):
    phi = random.uniform(0, 2*np.pi)
    theta = np.arccos(random.uniform(-1, 1))
    #theta = random.uniform(0, np.pi)
    particle_list.append([phi, theta])
particle_list = pd.DataFrame(particle_list, columns=['phi', 'theta'])   
```

```python
#plotting
fig = plt.figure()
plt.hist(particle_list['theta'], bins=50, histtype='step', color='r', label='phi')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Get the values of theta and phi from the particle_list dataframe
theta = particle_list['theta']
phi = particle_list['phi']

# Convert theta and phi to radians
theta_rad = np.radians(theta)
phi_rad = np.radians(phi)

# Convert spherical coordinates to Cartesian coordinates
x = np.sin(theta_rad) * np.cos(phi_rad)
y = np.sin(theta_rad) * np.sin(phi_rad)
z = np.cos(theta_rad)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y,z)

# Set labels for the axes
ax.set_xlabel('Phi')
ax.set_ylabel('Theta')
ax.set_zlabel('Count')

# Show the plot
plt.show()

```

```python
def isotrpic_sphere(n):
    """
    Generates coordinates on the surface of a sphere radius 1 uniformly
    
    params:
     - n (int)
        number of x, y and z points to be generated
    returns:
     - x, y, z (numpy.ndarrays)
        the coordinates on the surface of a sphere
    
    """
    u = np.random.uniform(size=n)
    theta = np.arccos(1-(2*u))

    phi = np.random.uniform(0, np.pi*2, n)
    # 2*np.arccos(1-(2*v)) doesnt work
    # massive hole in it
    
    r = 1
    x = r*np.sin(theta)*np.cos(phi)
    y = r*np.sin(theta)*np.sin(phi)
    z = r*np.cos(theta)
    

    return x, y, z


x, y, z = isotrpic_sphere(1000)


fig = plt.figure()
ax = Axes3D(fig)
# fix the scaling
ax.set_box_aspect((1,1,1))

ax.scatter(x, y, z, color='r')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
